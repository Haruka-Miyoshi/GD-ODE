```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

def f(x):
    return x**2

x=np.linspace(-10, 10, 100)
y=f(x)
plt.title('$y=x^2$')
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y)
plt.show()
```

```python
"""差分方程式"""
class Difference(object):
    def __init__(self):
        pass

    # 前進差分
    def forward(self, func, x0, x1):
        h=x1-x0
        df=func(x1)-func(x0)
        df/=h
        return df
    
    # 後退差分
    def backward(self, func, x0, x1):
        h=x0-x1
        df=func(x0)-func(x1)
        df/=h
        return df

"""最適化アルゴリズム:GD"""
class GradientDescent(object):
    """initialize"""
    def __init__(self, func):
        # 関数
        self.__func=func
        # 微分関数
        self.__ode=Difference()
        # 初期値x0
        self.__x0=np.random.randint(-10, 10)
        # Δxを設定
        self.__dx=0.01
        # λ学習係数
        self.__lambda=0.01
        # xの系列
        self.__x=[self.__x0]

    """最適化"""
    def optimization(self, epoch=1000):
        for i in range(epoch):
            z=-self.__lambda*self.__ode.forward(self.__func, self.__x[i], self.__dx)
            xi1=self.__x[i] + z
            self.__x.append(xi1)
        
        return self.__x

gd=GradientDescent(func=f)
xd=gd.optimization()
```

```python
yd=f(np.array(xd))
plt.title('$y=x^2$')
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y)
plt.scatter(xd,yd)
plt.show()
plt.show()
```
